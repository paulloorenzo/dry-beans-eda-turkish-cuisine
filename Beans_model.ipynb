### Feature Engineering

```python
beans_model['is_seker'] = beans_model['Class'] == 'SEKER'
beans_model['is_seker'].replace({True:'SEKER', False:'NON DERMASON'},
                               inplace=True)
```

```python
def custom_lambda(x):
    if x == 'SEKER' or x == 'HOROZ':
        return x
    
    return 'OTHERS'

beans_model['Class2'] = beans_model['Class'].apply(custom_lambda)\
     .astype('category').cat \
     .reorder_categories(['SEKER','HOROZ','OTHERS'])
```

## Exploratory Data Analysis

```python
beans_model.info()
```

```python
beans_model.drop(columns=['Class','is_seker','Class2']).corr()
```

```python
# Define the hexadecimal values for the classes
class2_colors = {
    'SEKER':'#2CA02C' ,
    'HOROZ':   '#9467BD'  ,
    'OTHERS':   '#607365' 
}

# Plotting
fig, ax = plt.subplots(figsize=(10, 6))
sns.histplot(data=beans_model, x='Area', hue='Class2',
             palette=class2_colors, bins=30, ax=ax)
ax.set_yscale('log')
ax.set_title('Histogram of Bean Areas between Seker and Non-seker')
ax.get_legend().set_title('Bean Type')
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(10,6))

sns.scatterplot(data=beans_model, x='MajorAxisLength', y='MinorAxisLength',
               hue='Class2', palette=['#2CA02C' ,'#9467BD' ,  '#607365' ],
               ax=ax,linewidths=0)
ax.get_legend().set_title('Bean Type')
ax.set_title('Bean Length by Bean Type')
ax.set_xlabel('Major Axis Length (mm)')
ax.set_ylabel('Minor Axis Length (mm)')

plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Define custom colors for each class
custom_colors = {'SEKER': '#2CA02C', 'BARBUNYA':  '#607365', 'BOMBAY':  '#607365', 'CALI': '#607365', 
                 'DERMASON':  '#607365', 'HOROZ': '#9467BD', 'SIRA': '#607365', 'CALIFORNIA':  '#607365'}

# Get the value counts of each class and sort them in descending order
class_counts = beans_model['Class'].value_counts().sort_values(ascending=False).index

# Create a count plot with the specified order and custom colors
plt.figure(figsize=(10, 6))
sns.countplot(data=beans_model, x='Class', palette=custom_colors, order=class_counts)

# Set labels and title
plt.xlabel('Bean Type')
plt.ylabel('Count')
plt.title('Total Value Counts of Each Class of Beans')

# Show the plot
plt.show()
```

```python
beans_model_subset = beans_model[beans_model['Class'].isin(['SEKER', 'HOROZ'])]
```

```python
fig, ax = plt.subplots(figsize=(8,4))

sns.violinplot(data=beans_model_subset, x='Solidity', y='Class',hue='Class', split=True,
               palette=['#2CA02C','#9467BD'], inner='quart', ax=ax)
ax.set_ylabel('Bean Type')
ax.set_xlabel('Solidity (px/shell)')
ax.set_title('Distribution of Bean Solidity (Seker vs Horoz)')

plt.show()
```
